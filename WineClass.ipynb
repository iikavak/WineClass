```python
q1, q3 = np.percentile(data['Malic acid'], [25, 75])

iqr = q3 - q1

lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

clean_data = data[(data['Malic acid'] >= lower_bound) 
                & (data['Malic acid'] <= upper_bound)]
q1, q3 = np.percentile(clean_data['Ash'], [25, 75])

iqr = q3 - q1

lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

clean_data = clean_data[(clean_data['Ash'] >= lower_bound) 
                        & (clean_data['Ash'] <= upper_bound)]
q1, q3 = np.percentile(clean_data['Alcalinity of ash'], [25, 75])

iqr = q3 - q1

lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

clean_data = clean_data[(clean_data['Alcalinity of ash'] >= lower_bound) 
                        & (clean_data['Alcalinity of ash'] <= upper_bound)]

q1, q3 = np.percentile(clean_data['Magnesium'], [25, 75])

iqr = q3 - q1

lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

clean_data = clean_data[(clean_data['Magnesium'] >= lower_bound) 
                        & (clean_data['Magnesium'] <= upper_bound)]

q1, q3 = np.percentile(clean_data['Proanthocyanins'], [25, 75])

iqr = q3 - q1

lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

clean_data = clean_data[(clean_data['Proanthocyanins'] >= lower_bound) 
                        & (clean_data['Proanthocyanins'] <= upper_bound)]

q1, q3 = np.percentile(clean_data['Color intensity'], [25, 75])

iqr = q3 - q1

lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

clean_data = clean_data[(clean_data['Color intensity'] >= lower_bound) 
                        & (clean_data['Color intensity'] <= upper_bound)]

q1, q3 = np.percentile(clean_data['Hue'], [25, 75])

iqr = q3 - q1

lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

clean_data = clean_data[(clean_data['Hue'] >= lower_bound) 
                        & (clean_data['Hue'] <= upper_bound)]

 
```

```python
fig, axs = plt.subplots(14,1,dpi=95, figsize=(10,26))
i = 0
for col in clean_data.columns:
    axs[i].boxplot(clean_data[col], vert=False)
    axs[i].set_ylabel(col)
    i+=1
plt.show()
```

```python
from scipy import stats
```

```python
from scipy.stats import entropy
```

```python
entropy(pk=clean_data)
```

```python
y= clean_data["class"].values
X= clean_data.drop("class",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.2,random_state=20)
```

```python
stats.zscore(X_train)
```

```python
stats.zscore(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn import metrics
```

```python
clf = DecisionTreeClassifier()
```

```python
clf = clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
clf = DecisionTreeClassifier(criterion="entropy", max_depth=20)
```

```python
clf = clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
